```python
import numpy as np
import scipy.misc
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
from scipy.ndimage.interpolation import rotate
```

```python
# x = np.loadtxt("train_x.csv", delimiter=",") # load from text 
x = np.load("x_centred_train.npy")
y = np.loadtxt("train_y.csv", delimiter=",") 
x = x.reshape(-1, 64*64) # reshape 
y = y.reshape(-1, 1) 
```

```python
neigh = NearestNeighbors(metric='euclidean')
```

```python
neigh.fit(x[:45000])
```

```python
answers = []
for i in range(45000, 50000):
    print("doing {0}".format(i))
    neighbors = neigh.kneighbors(x[i], 10)[1]
    
    classes = []
    for j in neighbors[0]:
        classes.append(y[j][0])
    answer = np.argmax(np.bincount(classes))
    answers.append(answer)
    

# neighbors = neigh.kneighbors(x[45000:50000],10)[1]
# predictions = []
# for n in neighbors:
#     classes = []
#     for i in n:
#         classes.append(y[i])
#     predictions.append(classes)

# predictions = [ [i[0] for i in p] for p in predictions]

# answers = [ np.argmax(np.bincount(p)) for p in predictions]
```

```python
comparison = ( answers == y[45000:50000].reshape(-1,))
correct_answers = np.sum(comparison + 0)
print(correct_answers)
```
